## 4. Data Understanding

### TimeStamp feature observations

* Our data is from 01-04-2018 to 31-08-2018. 
* Each date has 1440 observations, meaning our data is collected on a 1-minute frequency (24*60=1440)
* Total number of days in records are 153. (153*24*60 = 220,320 records)

---

### Missing Value information

* 11 rows have all the sensor values missing [2018-04-27 17:48:00 to 2018-04-27 17:55:00, 2018-07-11 09:12:00 to 2018-07-11 09:14:00]
* 8 rows have 50% values missing 


---


### Skewness and Kurtosis
* 84% of the sens0r features (42 nos) are significantly skewed (<-1 or >1)
* Out of the skewed sensor data (12 nos =>28.5%) are positively skewed (Right skewed)
* 78% of sensor data (39 nos) is having substantial kurtosis. (<-3 or >3)
* Positive kurtosis can be interpreted as heavily peaked and negative kurtosis can be understood as too flat.
* With kurtosis we can say that most features are having sudden impulsive records
* Kurtosis greated than 3 means Leptokurtic. Also means heavy tails, that is extreme outliers


---
### Standard Deviation
* We have very high standard deviation for most of the sensor features
* This could be due to substantial leptokurtic data indicating impulses/outliers
